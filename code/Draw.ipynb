```python
import matplotlib.pyplot as plt
import numpy as np
right=4
left=-right
x = np.arange(left, right, 0.01)
def initialize_plot():
    plt.xlim(left,right)
    plt.ylim(left,right)
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.spines['bottom'].set_position(('data',0))
    ax.spines['left'].set_position(('data',0))

name='sigmoid'
fig = plt.figure(figsize=(6,6))
initialize_plot()
y = 1.0/(1.0+np.exp(-x))
plt.plot(x, y, label = name)
plt.legend()
fig.savefig(name + ".pdf")
```

```python
name='tanh'
fig = plt.figure(figsize=(6,6))
initialize_plot()
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.plot(x, y, label = name)
plt.legend()
fig.savefig(name + ".pdf")
```

```python
fig = plt.figure(figsize=(6,6))
initialize_plot()
ReLU = np.where(x < 0, 0, x)
GeLU = 0.5*x*(1+ np.tanh(np.sqrt(2/np.pi)*(x+0.044715*np.power(x,3))))
plt.plot(x, ReLU, label = 'ReLU')
plt.plot(x, GeLU, label = 'GeLU')
plt.legend()
fig.savefig("relu_gelu.pdf")
```

```python
name='mse'
fig = plt.figure(figsize=(6,6))
initialize_plot()
y = x**2
plt.plot(x, y, label = name)
plt.legend([name])
fig.savefig(name + ".pdf")
```

```python
name='mae'
fig = plt.figure(figsize=(6,6))
initialize_plot()
y = abs(x)
plt.plot(x, y, label = name)
plt.legend([name])
fig.savefig(name + ".pdf")
```

```python
def huber_loss(x, d):
    return (abs(x)<=d)*x**2/2 + (abs(x)>d)*d*(abs(x)-d/2)

fig = plt.figure(figsize=(6,6))
initialize_plot()
plt.plot(x, x**2/2, label=r'$x^2/2$')
plt.plot(x, huber_loss(x, 1), label=r'huber: $\delta$=1')
plt.legend()
fig.savefig("huber.pdf")
```
